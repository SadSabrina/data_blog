# Функция распределения случайной величины, введение. 

```python
import matplotlib.pyplot as plt
import numpy as np


def get_distribution(variable):
    
    """Функция получает на вход дискретную переменную и отдает список списков, 
    с парой [величина, вероятность того, что эксперимент отдаст значение строго (!) меньшее, чем величина]
    
    Пример: переменная variable = [0, 1, 1, 2]
    Выход: [[0, 0.0], [1, 0.25], [2, 0.75], [3, 1.0]]
    
    """
    
    variable_unique_values = set(variable)
    variable_unique_values.add(max(variable) + 1)
    
    probas_distr = []
    
    for i in variable_unique_values:
        proba = 0
        
        for j in range(0, len(variable)):
            
            if variable[j] < i:
                
                proba += 1
        
        probas_distr.append([i, proba/len(variable)])  
        
            
    return probas_distr
            
        
```

#### Эксперимент 1. 

Эксперимент с подбрасыванием 2х монет. \
Случайная величина — количество выпавших орлов. 

Пространство исходов [ОО, ОР, РО, РР], [0 орлов, 1 орел, 1 орел, 2 орла]. 

```python
#Эксперимент с подбрасыванием 2х монет 
#Случайная величина — количество выпавших орлов


fig, ax = plt.subplots(figsize=(17, 10))
variable = [0, 1, 1, 2]
distr = get_distribution(variable)

X = [distr[i][0] for i in range(0, len(distr))]
Y = [distr[i][1] for i in range(0, len(distr))]


for i in range(0, len(distr)):
    
    plt.arrow(x=X[i], dx=-0.98, y=Y[i], dy=0,  width = 0.009)

        
   
plt.axvline(x=0, ymin=0.05, ymax=0.3, linestyle='--')
plt.axvline(x=1, ymin=0.3, ymax=0.75, linestyle='--')
plt.axvline(x=2, ymin=0.7, ymax=0.95,linestyle='--')


plt.title('Вероятностное распределение случайной величины', fontsize=20)
plt.ylabel('Вероятность того, что величина примет значение меньшее x', fontsize=16)
plt.yticks(np.arange(0, 1.05, 0.05))

plt.show()
```

#### Эксперимент 2. 

Моделирование вероятностей произвольного эксперимента в выборке из 10 испытаний. 

Исходы: \
2 единицы — вероятность 0.2 \
3 двойки — вероятность 0.3 \
4 тройки — вероятность 0.4 \
1 четверка — вероятность 0.1 

Случайная величина — диапазон чисел от 0 до 4. 

```python
variable = [2, 2, 2, 3, 4, 3, 1, 3, 3, 1]

distr = get_distribution(variable)


X = [distr[i][0] for i in range(0, len(distr))]
Y = [distr[i][1] for i in range(0, len(distr))]

fig, ax = plt.subplots(figsize=(17, 10))

for i in range(0, len(distr)):
    
    plt.arrow(x=X[i], dx=-0.98, y=Y[i], dy=0,  width = 0.009)


plt.axvline(x=1, ymin=0.05, ymax=0.25, linestyle='--')
plt.axvline(x=2, ymin=0.25, ymax=0.5, linestyle='--')
plt.axvline(x=3, ymin=0.5, ymax=0.87,linestyle='--')
plt.axvline(x=4, ymin=0.86, ymax=0.94,linestyle='--')

plt.title('Вероятность того, что величина примет значение меньшее x', fontsize=16)
plt.suptitle('Функция распределения случайной величины', fontsize=20)
plt.yticks(np.arange(0, 1.05, 0.1))

plt.show()
```

**Проверка:**

Вероятность величины **меньше** 1 равна 0.  \
Вероятность величины **меньше** 2 равна 0.2  \
Вероятность величины **меньше** 3 равна 0.2 + 0.3 равна 0.5 \
Вероятность величины **меньше** 4 равна 0.2 + 0.3 + 0.4 равна 0.9 \
Вероятность величины **меньше** 5 равна 1.
