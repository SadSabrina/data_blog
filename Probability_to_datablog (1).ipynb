# Испытание Бернулли. 

Вероятность k успехов в серии из n испытаний. 

```python
import matplotlib.pyplot as plt
from math import factorial
import numpy as np
```

```python
def probability_value(n, x, p, q):
    
    """
    n — число итераций эксперимента
    x — искоромое число успехов в серии из n испытаний
    p - вероятность успеха
    q — вероятность неудачи, q = 1 - p
    
    """
    
    binom_coef = factorial(n)/(factorial(x)*factorial(n-x))
    
    probab = p**x*q**(n-x)*binom_coef
    
    return probab

def probas_viz(probas):
    
    assert type(probas) == np.ndarray, 'Переменная должна иметь тип numpy.array'
    
    fig, ax = plt.subplots(figsize=(14, 8))
    
    ax.plot(probas[:, 0], probas[:, 1])
    
    maximum = probas[probas[:, 1] == probas[:, 1].max()]
    ax.plot(maximum[:, 0], maximum[:, 1], 'go')
    ax.text(maximum[:, 0], maximum[:, 1]+maximum[:, 1]*0.01, 'Maximum', fontsize=12)
    
    plt.xlabel('Количество успешных испытаний', fontsize=12)
    plt.ylabel('Вероятность стольких успешных испытаний', fontsize=12)
    ax.set_title(f'Вероятность k успехов в серии из {int(probas[:, 0].max())} испытаний', fontsize=15)
    
    plt.show()
    
    return 0
    
    
probas = np.array([[x, probability_value(7, x, 0.7, 0.3)] for x in range(1, 8)])
```

```python
#Задача про гречу

probas = np.array([[x, probability_value(7, x, 0.7, 0.3)] for x in range(1, 8)])
probas_viz(probas)
```

```python
#Задача про сон

probas = np.array([[x, probability_value(7, x, 0.6, 0.3)] for x in range(1, 8)])
probas_viz(probas)
```

```python
#А если увеличить количество... :)
#но об этом позже

probas = np.array([[x, probability_value(100, x, 0.6, 0.3)] for x in range(1, 101)])
probas_viz(probas)
```
