```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import random

random.seed(7)
```

```python
bimodal_distribution = np.concatenate([np.random.normal(loc=2, scale=0.5, size=100), 
                                       np.random.normal(loc=4, scale=0.5, size=100)])
```

```python
fig, ax = plt.subplots(figsize=(20, 6))

ax = sns.histplot(bimodal_distribution, bins=30, kde=True)
ax.set_title('Частный пример мультимодального распрделения — бимодальное распределение')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))

boxplot = ax1.boxplot(bimodal_distribution, patch_artist=True) #параметр patch_artist отвечает за полную окраску 
                                                            #боксплота
violinplot = ax2.violinplot(bimodal_distribution, showmedians=True)

boxplot['boxes'][0].set_color('lightblue')
violinplot['bodies'][0].set_color('lightblue')

ax1.set_title('Boxplot для бимодального распределения')
ax2.set_title('Violinplot для бимодального распределения')
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go


fig = make_subplots(1, 2)

fig.add_trace(go.Box(y=bimodal_distribution, name='Boxplot для бимодального распределения'), row=1, col=1)

fig.add_trace(go.Violin(y=bimodal_distribution, name='Violinplot для бимодального распределения'), row=1, col=2)

fig.update_layout(title_text='Различие наглядности boxplots и violinplots для мультимодальных распределений')
```
